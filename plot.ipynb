```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("band_structure.dat")
k_indices = data[:, 0]
energies = data[:, 1:]

for i in range(energies.shape[1]):
    plt.plot(k_indices, energies[:, i], color='black', lw=1)

# Optional: 라벨 위치 직접 지정
xticks = [0, 50, 100, 150]  # 예: M, Γ, K, M
xticklabels = ['M', 'Γ', 'K', 'M']
plt.xticks(xticks, xticklabels)

# plt.ylim(-20, 8)
plt.ylabel("Energy (eV)")
plt.xlabel("k-path")
plt.title("Band Structure")
plt.grid(True)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
