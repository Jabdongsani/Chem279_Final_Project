```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("band_structure_graphene.dat")
k_indices = data[:, 0]
energies = data[:, 1:]

plt.figure(figsize=(6, 8))

for i in range(energies.shape[1]):
    plt.plot(k_indices, energies[:, i], color='black', lw=1)

# Optional: 라벨 위치 직접 지정
xticks = [0, 50, 98, 153]  # 예: M, Γ, K, M
xticklabels = ['M', 'Γ', 'K', 'M']
plt.xticks(xticks, xticklabels)

plt.xlim(0, 153)
plt.ylim(-20, 8)
plt.ylabel("Energy (eV)")
plt.xlabel("k-path")
plt.title("Band Structure")
plt.grid(True)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt("band_structure_CNT.dat")

k_idx = data[:, 0]           # k-point index
energies = data[:, 1:]       # shape: (num_kpoints, num_bands)

# Plotting
plt.figure(figsize=(8, 6))
for band in range(energies.shape[1]):
    plt.plot(k_idx, energies[:, band], lw=1)
#k_labels = ['Γ', '', '', '', 'X']  # 원하는 위치에만 label
#plt.xticks([0, 25, 50, 75, 99], k_labels)
plt.xlim(0,20)
plt.ylim(-2,2)
plt.xlabel("k-point index")
plt.ylabel("Energy (eV)")
plt.axhline(y=0.0, color='gray', linestyle='--', lw=0.8)

plt.title("Band Structure of CNT")
plt.grid(True)
plt.tight_layout()
plt.savefig("cnt_bandstructure.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# Load data
data = np.loadtxt("band_structure_CNT_5_5.dat")
k_idx = data[:, 0]
energies = data[:, 1:]

# Plotting
plt.figure(figsize=(4, 6))
for band in range(energies.shape[1]):
    plt.plot(k_idx, energies[:, band], color='black', lw=0.8)

k_labels = ['Γ','X']  # 원하는 위치에만 label
plt.xticks([0, 25], k_labels)
plt.ylim(-2,2)
plt.xlim(0,30)
plt.axhline(y=0.0, color='gray', linestyle='--', lw=0.8)  # Fermi level
plt.xlabel("k-point index")
plt.ylabel("Energy (eV)")

plt.title("Band Structure of CNT (5, 5)")
plt.grid(True)
plt.tight_layout()
plt.savefig("cnt_bandstructure_smooth.png", dpi=300)
plt.show()
```
